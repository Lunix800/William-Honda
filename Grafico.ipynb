```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'./ESCOLAS 2014.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['ID_INTERNET', 'ID_BANDA_LARGA'])
df2 = pd.read_csv(r'./ESCOLAS 2015.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_INTERNET', 'IN_BANDA_LARGA'])
df3 = pd.read_csv(r'./ESCOLAS 2016.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_INTERNET', 'IN_BANDA_LARGA'])
df4 = pd.read_csv(r'./ESCOLAS 2017.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_INTERNET', 'IN_BANDA_LARGA'])
df5 = pd.read_csv(r'./ESCOLAS 2018.csv', delimiter = '|', encoding = 'iso-8859-1', usecols = ['IN_INTERNET', 'IN_BANDA_LARGA'])
```

```python
Dados2014 = []
Dados2015 = []
Dados2016 = []
Dados2017 = []
Dados2018 = []
anos= ['2014', '2015', '2016', '2017', '2018']
Dados = [Dados2014, Dados2015, Dados2016, Dados2017, Dados2018]
```

```python
Dados2014.append(df1.query('ID_INTERNET == 1' and 'ID_BANDA_LARGA == 1')['ID_INTERNET'].count())
Dados2015.append(df2.query('IN_INTERNET == 1' and 'IN_BANDA_LARGA == 1')['IN_INTERNET'].count())
Dados2016.append(df3.query('IN_INTERNET == 1' and 'IN_BANDA_LARGA == 1')['IN_INTERNET'].count())
Dados2017.append(df4.query('IN_INTERNET == 1' and 'IN_BANDA_LARGA == 1')['IN_INTERNET'].count())
Dados2018.append(df5.query('IN_INTERNET == 1' and 'IN_BANDA_LARGA == 1')['IN_INTERNET'].count())
```

```python
fig,ax = plt.subplots()
plt.rcParams['figure.figsize'] = (15,8)
plt.plot(anos, Dados, color='black')
plt.scatter(anos, Dados, color='red')
ax.plot(anos, Dados, label='CRESCIMENTO')
ax.grid()
plt.legend()
plt.title('Quantidade de escolas que possuem internet')
plt.ylabel('QUANTIDADE DE ESCOLAS')
plt.xlabel('ANOS')
plt.show()
```
